### Violin Grafik
**Veri Seti Hikayesi**

**total bill:** yemeğin toplam fiyatı (bahşiş ve vergi dahil)

**tip:** bahşiş

**sex:** ücreti ödeyen kişinin cinsiyeti (D=male, 1=female)

**smoker:** grupta sigara içen var mi? (0-No, 1=Yes)

**day:** gün (3-Thur, 4-Fri, 5-Sat, 6-Sun)

**time:** ne zaman? (0-Day, 1=Night)

**size:** grupta kaç kişi var?

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df
```

```python
sns.catplot(y = "total_bill", kind = "violin", data = df)
```

#### Violin Çaprazlamalar

```python
sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df)
```

```python
sns.catplot(x = "day", y = "total_bill", hue = "sex", kind = "violin", data = df)
```

## Korelasyon Grafikleri
### Scatter Plot
**Korelasyon:** Değişkenler arasında ilişkiyi temsil eden istatiksel verilerle oluşturulan denklem.
En çok saçılım grafiği kullanılmaktadır.

```python
sns.scatterplot(x = "total_bill", y = "tip", data =  df)
```

#### Scatter Plot Çaprazlama

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data =  df)
#kategorik değişken koyarak karar alınır.
```

```python
sns.scatterplot(x = "total_bill",y = "tip", hue = "time", style = "time", data =  df)
```

```python
sns.scatterplot(x = "total_bill",y = "tip", hue = "day", style = "time", data =  df)
#günler kırılımında bahşiş dağılımı gözlemlendi.
```

```python
#sürekli değişkeni boyut olarak eklersek:
#hue ile renklednrime yapma kararı alırsak
sns.scatterplot(x = "total_bill",y = "tip", hue = "size", size = "size", data =  df)
```

### Heat Map
* Uzun vadeli veriler (Zaman serisi, yıl, ay)
* Gözlenebilir zamansal verilerde
* Büyük ölcekli ve belirli periyotla olan olaylar

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df
```

```python
df["passengers"].describe()
#zamana bağlı yorumlamalar olduğunda trend göz önünde bulundurularak yprum yapılmalıdır.
#Yukarıdaki veride de hava taşımacııgı yıllar geçtikçe rağbet görmüş bu nedenle standart sapmadaki değişimler gözlemlenmiştir.
```

```python
#heatnmap bizden daha yapısal bir tablo bekliyor. bu nedenle aşağıdaki şekilde yaptığımız zaman hata almamız kaçınılmazdır.
sns.heatmap(df)
```

```python
df = df.pivot("month","year","passengers")
```

```python
df
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df, annot = True, fmt = "d")
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = .5)
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = .5, cbar = False)
```
